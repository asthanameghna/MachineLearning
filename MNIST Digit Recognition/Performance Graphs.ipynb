#### 4. Comparison with SKLearn Models

```python
n_groups = 4
mse = [1.36, 143.19, 159.66, 254.74]
mse_sk = [9.52, 143.34, 12.14, 10.00]
models = ['LR','NN','RF','GP']
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, mse, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Our Model')
 
rects2 = plt.bar(index + bar_width, mse_sk, bar_width,
                 alpha=opacity,
                 color='g',
                 label='SKLearn Model')
 
plt.xlabel('Models')
plt.ylabel('Mean Squared Error')
plt.title('MSE Comparison')
plt.xticks(index + bar_width, ('LR', 'NN', 'RF', 'GP'))
plt.legend()
 
plt.tight_layout()
plt.show()
```
